Part 3: Combining Concepts

```python
full_name=input("Enter your name:")
```

```python
letter_list=list(full_name.replace(" ",""))
```

```python
letter_list
```

```python
for i in range(len(letter_list)):
    if i<3:
        letter_list[i]=letter_list[i].upper()
    else:
        letter_list[i]=letter_list[i].lower()
        
```

```python
letter_list
```

```python
name="".join(letter_list)
```
