```python
# 1-numbers from 1 to 50 

for i in range(1,51):
    print(i)
```

```python
# 2-numbers between 1 and 100

for i in range(1,100):
    if i%2 == 0:
        print(i)
```

```python
# 3- sum of numbers from 1 to N
num=int(input("Enter a number:"))
sum=0
for i in range(1,num+1):
    sum=sum+i
print("Sum of number is:",sum)
    
```

```python
# 4- multiplication table
n=int(input("Enter number to create table:"))
for i in range(1,11):
    print(f"{n}x{i}={n*i}")
    
    
```

```python
# 5- Reverse string
string=str(input("Enter a string:"))
reverse_s=""
for letter in string:
    reverse_s=letter + reverse_s
print("Reverse_string:",reverse_s)
```

```python
# 6- Count Vowels

vowels=['a','e','i','o','u']
count=0
string=str(input("Enter a string"))
for i in string:
    if i in vowels:
        count=count+1
print("Number of vowels:",count)
```

```python
# 7- pattern

for i in range(1,6):
    print("*"*i)
```

```python
# 8- Prime number

number=int(input("Enter a number:"))
if number<=1:
    print("Not prime!")
else:
    for i in range(2,number):
        if number % i==0:
            print("Not prime!")
            break
    else:
        print("Prime Number!")
```

```python
# 9- find divisible of 3

numbers = [12, 5, 7, 18, 21, 30]
for i in numbers:
    if i % 3==0:
        print(i)
```

```python
# 10- Break and continue

for i in range(1,21):
    if i % 3==0:
        continue
    elif i>15:
        break
    print(i)
```

```python
# 11- Facto
number=int(input("Enter a number:"))
num=number
fact=1
while number>0:
    fact=fact*number
    number-=1
print(f"Factorial of {num} is: {fact}")
    
```
