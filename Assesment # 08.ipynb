```python
# Grade_Calculations

def calculate_grade(marks):
    if marks >=90:
        print("A")
    elif 80 <= marks <90:
        print("B")
    elif 70 <= marks <80:
        print("C")
    elif marks < 70:
        print("Fail")
```

```python
calculate_grade(66)
```

```python
# Even_Number

def is_even(number):
    if number%2==0:
        return True  
    else:
        return False
```

```python
is_even(6)
```

```python
# Calculate Factorial

def factorial(n):
    fac=1
    for i in range(1,n+1):
        fac=fac*i
    return fac 
```

```python
factorial(3)
```

```python
#Find Maximum Number

def find_max(a,b,c):
    if a >=b and b>=c:
        print(a)
    elif b>=a and b>=c:
        print(b)
    else:
        print(c)
```

```python
find_max(3,10,10)
```

```python
# Word Counter

def count_words(sentence):
    spaces=sentence.count(" ")
    words_count=spaces+1
    print("Number of words:",words_count)
    

    
```

```python
count_words("i love")
```

```python
# Prime Number Checker

def is_prime(num):
    if num<=1:
        print("Not prime")
    for i in range(2,num):
        if num % i==0:
            print(" Not a prime number!")
            break
    else:
        print("prime number!")
    
```

```python
is_prime(4)
```

```python
#  List Sum Function

def list_sum(numbers):
    sum=0
    for i in numbers:
        sum=sum+ i
    print("Sum of list=",sum)
        
```

```python
list_sum([1,2,3,5])
```

```python
#Password Strength Checker

def check_password(password):
    if len(password) < 8:
        return "Weak"
    if any(ch.isdigit() for ch in password):
        return "Strong"
    return "Weak"
```

```python
check_password("123467ghi54")
```

```python
#Temperature Converter

def celsius_to_fahrenheit(c):
    F = ((c * 9/5) + 32)
    return F
```

```python
celsius_to_fahrenheit(55)
```

```python

def remove_duplicates(data):
    n_list=set(data)
    return list(n_list)
```
