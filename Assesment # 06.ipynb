```python
availible_cash=50
item_price=float(input("Enter purchased item price:"))
tax=(3/100)*item_price
total_pay=item_price+tax
print("Item_price=",total_pay)
cash_left=availible_cash-total_pay
print("Money left after purchasing item: ",cash_left,"$")
```

```python
number=int(input("Enter your marks:"))
if 90<=number <=100:
    print("A Grade")
elif 80<=number <=89:
    print("B Grade")
elif 70<=number <=79:
    print("C Grade")
elif 60<=number <=69:
    print("D Grade")
elif 0<=number <=59:
    print("F Grade")
else:
    print("Invalid marks")
```
